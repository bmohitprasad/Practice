```python
year = 1904

if (year % 4 == 0):
    if (year % 100 == 0):
        if (year % 400 == 0):
            print("Leap Year")
        else:
            print("Not a Leap Year")
    else:
        print("Leap Year")
else: 
    print("Not a Leap Year")
```

```python
year = 1900

print("Leap year" if (year % 4 == 0 and (year % 100 != 0 or year % 400 == 0)) else "Not a leap year")
```

```python
print("Leap year" if (year % 400 == 0 or (year % 100 != 0 and year % 4 == 0)) else "Not a leap year")
```
