##### list comprehension

```python
l = [1, 2, 3, 4, 5, 6, 7]

l2 = [n for n in l if (n % 2 == 0)]
print (l2)
```

```python
def is_pass (marks):
    return any([True for m in marks if m < 35])

print ("Fail" if is_pass([36, 35, 40]) else "Pass")
```

##### Perfect Square or not


```python
import math

n = float(input("Enter the number: "))
n1 = float(int(math.sqrt(n)))
n2 = math.sqrt(n)


print("Perfect square" if n1 == n2 else "Not a Perfect square")
```
