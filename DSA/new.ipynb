#### Syntax:- true_value IF condition ELSE false_value


```python
def is_even(num):
    return num% 2 == 0

res = is_even(5)
# print(res)
print("Even" if is_even(5) else "Odd")

# Syntax:- true_value IF condition ELSE false_value
```

```python
def is_divisible (num):
    return (num % 5 == 0 and num % 10 != 0)
    


print("Satisfied" if is_divisible (15) else "Not Satisfied")
```

```python
def is_pass(marks: list):
    return any( # any checks if there is any true value in the list
        [m < 35 for m in marks] # Puts false/true into a list
    ) 

print("Fail" if is_pass([75, 80, 34]) else "Pass")
    
# return m for m in marks
```

##### Clear any one subject , PASS

```python

def is_pass(marks: list):
    return any([m > 35 for m in marks]) 

print("Pass" if is_pass([20, 38, 25]) else "Fail")
```

```python

def is_passs(marks: list):
    count = 0
    for m in marks:
        if m > 35: count += 1
    
    return count > 0

print ("Pass" if is_passs([20, 38, 25]) else "Fail")
```

##### Clear any two subject , PASS

```python

def is_pass(marks: list):
    count = 0
    for m in marks:
        if m > 35: count += 1
    
    return count > 1

print ("Pass" if is_passs([20, 38, 25]) else "Fail")
```
