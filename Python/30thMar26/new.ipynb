####
1. w+ ==> write, read

```python

with open('file2.txt', 'w+') as h:
    h.write('''Lorem ipsum dolor, sit amet consectetur adipisicing elit. Nulla ipsum amet cumque voluptatem
reprehenderit, saepe consectetur vitae porro dolorem repellendus dolor maiores iure non rem a,
voluptate maxime quasi voluptatibus.''')

    h.seek(0) # change cursor position
    # h.seek(10)
    print(h.tell()) # gets cursor position
    data = h.read()
    print(data)
```

#### 
2. r+ ==> read, write

```python
with open('file2.txt', 'r+') as h:
    data = h.read()
    print(data)
```

```python
with open('file2.txt', 'r+') as h:
    data = h.read()
    h.write("1234353451")
```

```python
with open('file2.txt', 'r+') as h:
    h.write("1234353451 57645784386")
```

```python
with open('file2.txt', 'r+') as h:
    h.seek(30)
    h.write("57645784386")
```

#####
3. a+ ==> append, read

```python
with open ('file2.txt', 'a+') as v:
    v.write("Hello from a+")
    
    v.seek(0)
    data = v.read()
    print(data)
```

```python
r = open('file3.txt', 'r')

print("No. of characters: ", len(r.read()))
```

##### No. of words

```python
r = open('file3.txt', 'r')

lines = r.readlines()
count = 0
# print(lines)
# print(len(lines))
for i in range (0, len(lines)):
    line = lines[i]
    count += len(line.split(' '))

print("No. of words: ", count)
```
