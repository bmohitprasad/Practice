```python
import MySQLdb
import random

try: 
    con = MySQLdb.connect(user='root', host='127.0.0.1', database='PDBC_84_db', password='738107')
    cur = con.cursor()
    
    q = '''CREATE TABLE IF NOT EXISTS BANK_SBI (
    ACCOUNT_NUMBER BIGINT PRIMARY KEY,
    HOLDER_NAME VARCHAR(50),
    IFSC_CODE VARCHAR(20),
    ACCOUNT_TYPE VARCHAR(20),
    MOBILE BIGINT,
    BALANCE FLOAT
    )'''
    cur.execute(q)
    con.commit()

except MySQLdb.Error as e:
    print(f"Database Error: {e}")

class SBI:
    def create_account(self):
        acc_name = input("Enter account holder name: ")
        ifsc = "SBI0123"
        acc_type = input("Enter account type (savings / zero): ").lower()
        mobile = int(input("Enter mobile number: "))
        acc_num = random.randint(1000000000, 9999999999)

        balance = 1000 if acc_type == "savings" else 500

        try:
            # 2. Use the existing global cursor
            q = "INSERT INTO BANK_SBI VALUES (%s, %s, %s, %s, %s, %s)"
            data = (acc_num, acc_name, ifsc, acc_type, mobile, balance)

            # 3. Pass data as the second argument!
            cur.execute(q, data)
            con.commit()
            print(f"Account created successfully! Account Number: {acc_num}")
            
        except MySQLdb.Error as e:
            print(f"Error during insertion: {e}")
            con.rollback() # Good practice to undo changes if something fails

obj = SBI()
obj.create_account()
```
