```python
class student:
    '''Student demo class'''
    def __init__(self):
        self.name = "Mohit"
        self.age = 22
        self.marks = 80

    def display(self):
        print("Student name: {}\nStudent's age: {}\nStudent's marks: {}".format(self.name, self.age, self.marks))

obj = student()
print(obj.__doc__)

obj.display()

print(obj.__dict__)
```

#### Declare Instance variables

```python
class Test:
    def __init__(self):
        self.a = 10
        self.b = 30

obj = Test()
print(obj.__dict__)
```

```python
class Test:
    def __init__(self):
        self.a = 10
        self.b = 30

    def method (self):
        self.c = 20

obj = Test()
obj.method()

print(obj.__dict__)
```

```python
class Test:
    def __init__(self):
        self.a = 10
        self.b = 30

obj = Test()
obj.c = 40

print(obj.__dict__)
```

#### Accessing variables inside instance methods and outside class

```python
class Test:
    def __init__(self):
        self.a = 10
        self.b = 20

    def method (self):
        print("Inside the class")
        print(self.a)
        print(self.b)

obj = Test()

obj.method()

print ("Outside the class")
print ("a: ", obj.a)
print ("b: ", obj.b)
```

#### Deleting variables inside and outside the class

```python
class employee:
    def __init__(self):
        self.name = "Mohit"
        self.age = 22
        self.salary = 20000
        self.gender = "male"

    def method (self):
        del self.gender

emp = employee()
print(emp.__dict__)

emp.method()
print(emp.__dict__)

del emp.salary
print(emp.__dict__)

```
