```python
class student:
    '''Student demo class'''
    def __init__(self):
        self.name = "Mohit"
        self.age = 22
        self.marks = 80

    def display(self):
        print("Student name: {}\nStudent's age: {}\nStudent's marks: {}".format(self.name, self.age, self.marks))

obj = student()
print(obj.__doc__)

obj.display()

print(obj.__dict__)
```

#### Accessing variables inside instance methods and outside class

```python
class Test:
    def __init__(self):
        self.a = 10
        self.b = 20

    def method (self):
        print("Inside the class")
        print(self.a)
        print(self.b)

obj = Test()

obj.method()

print ("Outside the class")
print ("a: ", obj.a)
print ("b: ", obj.b)
```

#### Deleting variables inside and outside the class

```python
class employee:
    def __init__(self):
        self.name = "Mohit"
        self.age = 22
        self.salary = 20000
        self.gender = "male"

    def method (self):
        del self.gender

emp = employee()
print(emp.__dict__)

emp.method()
print(emp.__dict__)

del emp.salary
print(emp.__dict__)

```

### Class/Static variables


```python
class Test:
    x = 100
    def __init__(self):
        self.y = 99

t1 = Test()
t2 = Test()

print(Test.x)
print(t1.__dict__)

print("t1 :", t1.x, t1.y)
print("t2 :", t2.x, t2.y)

Test.x = 1000
t1.y = 999

print("t1 :", t1.x, t1.y)
print("t2 :", t2.x, t2.y)

```

#### Various places to declare static variables:
1. In general we can declare within the class directly but from out side of any method
2. Inside constructor by using class name
3. Inside instance method by using class name
4. Inside classmethod by using either class name or cls variable
5. Inside static method by using class name|

```python
class Test: 
    a = 10

    def __init__(self):
        Test.b = 20

    def m1 (self):
        Test.c = 30 
    
    @classmethod
    def m2(cls):
        cls.d1 = 40
        Test.d2 = 50

    @staticmethod
    def m3():
        Test.e = 60

print(Test.__dict__)

t1 = Test()
print(Test.__dict__)

t1.m1()
print(Test.__dict__)

t1.m2()
print(Test.__dict__)

t1.m3()
print(Test.__dict__)

Test.f = 60
print(Test.__dict__)
```

#### How to access static variables:
1. inside constructor: by using either self or classname
2. inside instance method: by using either self or classname
3. inside class method: by using either cls variable or classname
4. inside static method: by using classname
5. From outside of class: by using either object reference or classnmae

```python
class Test:
    a = 10
    def __init__(self):
        print("From Constructor:", self.a)
        print("From Constructor:", Test.a)

    def m1 (self):
        print("From instance method:", self.a)
        print("From instance method:", Test.a)

    @classmethod
    def m2 (cls):
        print("From class method:", cls.a)
        print("From class method:", Test.a)

    @staticmethod
    def m3 ():
        print("From static method", Test.a)

print(Test.__dict__)
print(Test.a)

t1 = Test()

print("Outside the class: ", Test.a)
print("Outside the class: ",t1.a)

t1.m1()
t1.m2()
t1.m3()

```

#### Updating static variable

```python
class Test:
    a = 666
    def __init__(self):

        print("INSIDE CONSTRUCTOR")
        print()
        print("1. Before change Test.a: ", Test.a)
        Test.a = 777

        print("After: ")
        print(Test.a)

        print("2. Before change self.a: ", self.a)

        self.a = 778
        print("After: ")
        print(self.a)

        print("====================================")

    def method (self):
        print("INSIDE INSTANCE METHOD")


        print("3. Before change Test.a: ", Test.a)

        Test.a = 888
        print("After: ")

        print(Test.a)

        print("4. Before change self.a: ", self.a)

        self.a = 889
        print("After: ")

        print(self.a)

        print("====================================")


    def method4 (self):
        print("INSIDE ANOTHER INSTANCE METHOD")


        print("5. Before change Test.a: ", Test.a)

        Test.a = 890
        print("After: ")

        print(Test.a)

        print("6. Before change self.a: ", self.a)

        self.a = 891
        print("After: ")

        print(self.a)

        print("====================================")


    @classmethod
    def method2 (cls):
        print("INSIDE CLASS METHOD")

        print("7. Before change cls.a: ", cls.a)

        cls.a = 892
        print("After: ")

        print(cls.a)

        print("8. Before change Test.a: ", Test.a)

        Test.a = 893
        print("After: ")

        print(Test.a)
        print("====================================")


    @staticmethod
    def method3 ():
        print("INSIDE STATIC METHOD")

        print("9. Before change Test.a: ", Test.a)

        Test.a = 999
        print("After: ")

        print(Test.a)


print(Test.a)

Test.a = 667

print("BEFORE OBJECT CREATION :",Test.a)
print("++++++++++++++++++++++++++++++++++++++")

t = Test()

print("AFTER OBJECT CREATION :",Test.a)
print("++++++++++++++++++++++++++++++++++++++")
t.a
t.method()
t.method4()
t.method2()
t.method3()
```
