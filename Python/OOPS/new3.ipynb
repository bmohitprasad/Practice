#### Super keyword

```python
class Parent:
    def __init__(self):
        print("Parent class")

class Child (Parent):
    def __init__(self):
        super().__init__()
        print("Child class")

obj = Child()

# Constructor overriding
```

```python
class Parent:
    def __init__(self):
        print("Parent class")

class Child (Parent):
    def method (self):
        print("Hello from inst_method of child class")

objC = Child()
objC.method()
```

```python
class bank:
    def __init__(self, acc_no):
        self.acc_no = acc_no
        print("Hello from Parent class Bank")

class SBI(bank):
    def __init__(self, bank_name, acc_no):
        self.bank_name = bank_name
        super().__init__(acc_no)
        print(self.bank_name)
        print(self.acc_no)

obj = SBI("SBI", 101)
```

##### Single level inheritence : (One to One)

```python
class admin: 
    def __init__(self, admin_name, admin_id, admin_salary):
        self.admin_name = admin_name
        self.admin_id = admin_id
        self.admin_salary = admin_salary

class user (admin):
    def __init__(self, admin_name, admin_id, admin_salary, user_name, user_id, user_salary):
        self.user_name = user_name
        self.user_id = user_id
        self.user_salary = user_salary

        super().__init__(admin_name, admin_id, admin_salary)

        print("========== user details ==========")

        print("User name: ", self.user_name)
        print("User id: ", self.user_id)
        print("User salary: ", self.user_salary)

        print("========== Admin details ==========")
        
        print("Admin name: ", self.admin_name)
        print("Admin id: ", self.admin_id)
        print("Admin salary: ", self.admin_salary)

obj = user("Mohit", 101, 30000, "Prem", 201, 20000)
```

##### Multiple inheritence : (Many to One)

```python
class parent1:
    def __init__(self):
        self.p1_name = "Mohit"
        self.p1_age = 22

class parent2: 
    def p2_method(self):
        self.p2_name = "Prem"
        self.p2_age = 20

class child (parent1, parent2):
    def child_method(self):
        self.name = "Bunny"
        self.age = 1
    
    def details(self):
        print("Parent 1 details")
        print(self.p1_name)
        print(self.p1_age)

        print("Parent 2 details")
        print(self.p2_name)
        print(self.p2_age)

        print("Child details")
        print(self.name)
        print(self.age)

obj = child()
obj.p2_method()
obj.child_method()
obj.details()
```

##### Multilevel Inheritence

```python
class grandparent:
    def __init__(self):
        self.g_name = "Rajesh"
        self.g_age = 80

class parent (grandparent):
    def f_method(self):
        self.f_name = "Hari"
        self.f_age = 30

class child (parent):
    def child_method(self):
        self.c_name = "Ansh"
        self.c_age = 5

    def display(self):
        print("=========== Grandparent ===========")
        print(self.g_name)
        print(self.g_age)

        print("=========== Parent ===========")
        print(self.f_name)
        print(self.f_age)

        print("=========== Child ===========")
        print(self.c_name)
        print(self.c_age)

obj = child()
obj.f_method()
obj.child_method()
obj.display()
```

##### Hierarchical Inheritance

```python
class parent:
    def __init__(self):
        self.p_name = "Mohit"
        self.p_age = 50

class child1 (parent):
    def child1_method (self):
        self.ch1_name = "Ch1"
        self.ch1_age = 30

class child2 (parent):
    def child2_method (self):
        self.ch2_name = "Ch2"
        self.ch2_age = 25

class child3 (parent):
    def child3_method (self):
        self.ch3_name = "Ch3"
        self.ch1_age = 20

obj1 = child1 ()
obj1.child1_method()

obj2 = child2 ()

obj2.child2_method()

print(obj1.ch1_name)
print(obj1.ch1_age)

print(obj1.p_name)
print(obj1.p_age)

print(obj2.ch2_age)
```

#### Hybrid Inheritance

```python
class gfclass:
    def gf_method(self):
        self.gf_name = "Rajesh"
        self.gf_age = 50

class parent1 (gfclass):
    def p1_method(self):
        self.p1_name = "Mohit"
        self.p1_age = 22

class parent2 (gfclass): 
    def p2_method(self):
        self.p2_name = "Prem"
        self.p2_age = 20

class child (parent1, parent2):
    def child_method(self):
        self.name = "Bunny"
        self.age = 1
    
    def details(self):
        print("Grandparent details")
        print(self.gf_name)
        print(self.gf_age)

        print("Parent 1 details")
        print(self.p1_name)
        print(self.p1_age)

        print("Parent 2 details")
        print(self.p2_name)
        print(self.p2_age)

        print("Child details")
        print(self.name)
        print(self.age)

obj = child()
obj.gf_method()
obj.p1_method()
obj.p2_method()
obj.child_method()
obj.details()
```
