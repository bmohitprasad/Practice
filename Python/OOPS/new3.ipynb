#### Super keyword

```python
class Parent:
    def __init__(self):
        print("Parent class")

class Child (Parent):
    def __init__(self):
        super().__init__()
        print("Child class")

obj = Child()

# Constructor overriding
```

```python
class Parent:
    def __init__(self):
        print("Parent class")

class Child (Parent):
    def method (self):
        print("Hello from inst_method of child class")

objC = Child()
objC.method()
```

```python
class bank:
    def __init__(self, acc_no):
        self.acc_no = acc_no
        print("Hello from Parent class Bank")

class SBI(bank):
    def __init__(self, bank_name, acc_no):
        self.bank_name = bank_name
        super().__init__(acc_no)
        print(self.bank_name)
        print(self.acc_no)

obj = SBI("SBI", 101)
```
