####    Encapsulation

##### Public variables


```python
class employee:
    emp_id = 101
    def __init__(self):
        self.emp_name = "emp_name"

    def details (self):  
        print("Employee name: ", self.emp_name)
        print("Employee id: ", employee.emp_id)
    
class depart (employee):
    def dep_method(self):
        print("Employee name: ", self.emp_name)
        print("Employee id: ", employee.emp_id)


emp = depart()
emp.details()
emp.dep_method()
```

##### Private variables

```python
class school:
    def __init__(self):
        self.__school_name = "De Paul"

    # def details(self):
    #     print("School name: ", self.__school_name)

class teachers (school):
    def details (self):
        print(self.__school_name)

obj = teachers()
obj.details()
```

```python
class school:
    def __init__(self):
        self.__school_name = "De Paul"

    def details(self):
        print("School name: ", self.__school_name)

obj = school()
obj.details()
```

##### private variables are only accessible in the same class 

```python
class bank:
    __balance = 10000

    def __init__(self):
        print("Bank balance: ", self.__balance)

obj = bank()
```

##### Aceessing private variables outside the class

```python
class bank:
    __balance = 10000

    def __init__(self):
        self.__holder_name = "sai"
        bank.__balance += 2000

    def display (self):
        print("Updated balance: ", bank.__balance)

obj = bank()
obj.display()

# Name mangling
print(obj._bank__balance)
print(obj._bank__holder_name)
```

##### Protected

```python
class employee:
    _salary = 20000
    def __init__(self):
        self._emp_id = 101

class salesDept (employee):
    def details(self):
        print(self._emp_id)

obj = salesDept()
obj.details()
print(obj._salary)
```

```python
class BankAccount:
    def __init__(self, name, balance):
        self.name = name              # Public
        self._account_type = "Debit"  # Protected
        self.__balance = balance      # Private

    def display(self):
        # All are accessible inside the class
        print(f"Owner: {self.name}, Type: {self._account_type}, Balance: {self.__balance}")

acc = BankAccount("Alice", 1000)

print(acc.name)           # Works: Alice
print(acc._account_type)  # Works: (But discouraged)
# print(acc.__balance)    # Throws AttributeError
```
