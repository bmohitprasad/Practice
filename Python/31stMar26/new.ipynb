#### Decorators

```python
def Decorators(wish):
    def inner():
        print('before function')
        wish()
        print('after calling')
    
    return inner

@Decorators
def wish():
    print('good morning')

wish()
```

```python
def even(a):
    def inner():
        for i in range (10, 30):
            if i % 2 == 0:
                print(i)
    return inner

@even
def fun():
    for x in range (10, 30):
        print(x)

fun()
```

```python
def prime(fun):
    def inner():
        for x in range (10, 20): 
            for y in range (2, x):
                if (x % y == 0): 
                    break
            
            else: 
                print(x)
    return inner

@prime
def fun():
    for x in range (10, 30):
        print(x)

fun()
```

```python
def even_num(check_num):
    def inner(b):
        if b % 2 == 0:
            print("Even")
        else: 
            print("Odd")
    return inner

@even_num
def check_num(a):
    print(a)

check_num(11)
```

```python
def prime_nums(func):
    def inner(a, b):
        for x in range(a, b):
            for y in range (2, x):
                if (x % y == 0):
                    break
            else:
                print(x)

        func(a, b)
    return inner

@prime_nums
def num (a, b):
    for x in range(a, b):
        print(x)

num(25, 35)
```

```python
def authentication(log):
    def inner(usrName, pswd):
        if (usrName == "Mohit"):
            if (pswd == "Mohit@123"):
                print('valid user, Logging in....')
            else:
                print("password not matched")
        else:
            print("invalid username")
    return inner


@authentication
def log (username, password):
    print(username)
    print(password)

log("Mohit", "Mohit@123")
```

##### Multiple Decorators

```python
def fun1(func):
    def inner():
        print("Decorator1")
        func()
    return inner

def fun2(func):
    def inner():
        print("Decorator2")
        func()
    return inner

@fun1
@fun2
def fun():
    print("Multiple decorators")

fun()
```
