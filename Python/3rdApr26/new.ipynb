#### Generators

```python
res = (x for x in range (100)) # tuple comprehension

print (res)
print (type(res))
print (next(res))
print (next(res))

# print all values of the generator
for x in res:
    print(x)

# indexing not allowed 
```

```python
def num():
    for x in range(10):
        if x % 2 == 0:
            yield x # for generator

print(next(num()))
obj = num()

print(next(obj))
print(next(obj))
```

```python
def prime (a, b):
    for x in range (a, b):
        for y in range (2, x):
            if x % y == 0:
                break
        else:
            yield x # if given return, then function, if yield then generator

obj = prime(20, 30)
# print(next(obj))

# for x in obj:
#     print(x)


for x in prime(20, 30):
    print (x)
```

```python
def fileOps():
    with open ('file.txt', 'r') as f:
        data = f.read()
        yield data

obj = fileOps()
print (next(obj))

```

```python
def fibo(r):
    a = 0
    b = 1

    for x in range (r):
        if (x == 0):
            yield a
        
        elif (x == 1):
            yield b

        else:
            c = a + b
            a = b
            b = c
            yield c

obj = fibo (10)
print (next(obj))
print (next(obj))
print (next(obj))
print (next(obj))
print (next(obj))
print (next(obj))
```

```python
def fibo (r):
    a = 0
    b = 1
    for x in range (r):
        yield a
        a, b = b, a + b

obj = fibo (10)

for x in obj:
    print(x)
```
