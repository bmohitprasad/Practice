```python
con = MySQLdb.connect(user = 'root', host = '127.0.0.1', database = 'pdbc_84', password = '738107')
cur = con.cursor()

n = int(input("Enter number of records: "))

for i in range (n):
    id = int(input(f"Enter pin of employee {i}: "))
    name = input(f"Enter name of employee {i}: ")
    salary = int(input(f"Enter salary of employee {i}: "))

    q = f"INSERT INTO EMPLOYEE VALUES ({id}, '{name}', {salary})"
    cur.execute(q)
    con.commit()

con.close()
```
