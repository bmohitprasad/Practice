```python
con = MySQLdb.connect(user = 'root', host = '127.0.0.1', database = 'pdbc_84', password = '738107')
cur = con.cursor()

n = int(input("Enter number of records: "))

for i in range (n):
    id = int(input(f"Enter pin of employee {i + 1}: "))
    name = input(f"Enter name of employee {i + 1}: ")
    salary = int(input(f"Enter salary of employee {i + 1}: "))

    q = f"INSERT INTO EMPLOYEE VALUES ({id}, '{name}', {salary})"
    cur.execute(q)
    con.commit()

con.close()
```

##### Using while loop


```python
con = MySQLdb.connect(user = 'root', host = '127.0.0.1', database = 'pdbc_84', password = '738107')
cur = con.cursor()

while True:
    choice = int(input('''
    1. Insert data
    2. Exit
    '''))

    if choice == 2:
        break
    
    elif choice != 2 or choice != 1: 
        print("Enter valid choice (1 or 2): ")

    else: 
        id = int(input(f"Enter pin: "))
        name = input(f"Enter name: ")
        salary = int(input(f"Enter salary: "))

        q = f"INSERT INTO EMPLOYEE VALUES ({id}, '{name}', {salary})"
        cur.execute(q)
        con.commit()

con.close()
```

##### Multiple records

```python
import MySQLdb
con = MySQLdb.connect(user = "root", host = "127.0.0.1", database = "pdbc_84", password = "738107")
cur = con.cursor()

q = "INSERT INTO EMPLOYEE VALUES (%s, %s, %s)"

insert_data = [(6, 'Afroz', 15000), (7, 'Rohit', 10000), (8, 'Sain', 25000)]
cur.executemany(q, insert_data)
con.commit()
con.close
```
