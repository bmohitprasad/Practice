### File Handling


#### Functions
1. open
2. read
3. write
4. close

#### Modes
1. x ==> create new file, write
2. w ==> file create, write
3. r ==> only read
4. a ==> creat, append

```python
s = open('file1.html', 'x')
```

```python
s1 = open('style.css', 'x')
s2 = open('db.sql', 'x')
s3 = open('file.txt', 'x')
s4 = open('python.py', 'x')

```

```python
n = open('new1.py', 'x')
n.write('v = 1000')
n.close
```

```python
o = open('file2.txt', 'w')
o.write('Files are very common permanent storage areas to store our data.')
o.close()
```

```python
k = open('file3.txt', 'w')

for i in range (20, 41, 2):
    k.write(f'{i} ')

k.close()
```

```python
k = open('prime.txt', 'w')

for i in range (20, 40):
    for j in range (2, i):
        if (i % j == 0):
            break
    else:
        k.write(f'{i}\n')


k.close()
```

##### Read

```python
r = open ('new1.py', 'r')
data = r.read()
print(data)
```

Append


```python
r = open ('prime.txt', 'a')
r.write("\nNew data")
```

```python
r = open ("new1.py", 'r')
data = r.read()
print(type(data))

new = open ("new3.py", 'w')
new.write(data)

r.close()
new.close()
```

##### No need to close the file using with

```python
with open ('new.py', 'r') as h:
    print(h.read())
```
